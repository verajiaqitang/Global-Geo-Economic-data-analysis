```python
import geopandas

import geopandas as gpd
import pandas as pd
```

```python
countrylist=["Latvia"]
```

```python
countrylist=["Bolivia","Bulgaria","Croatia","Czech","Dominican","Estonia","Finland","Iceland","India",
             "Indonesia","Israel"
             ,"Italy","Macedonia","Moldova","Montenegro","Peru",
             "Poland","Romania","Russia","Serbia","Slovakia","SouthAfrica","Spain","Switzerland"]
```

```python
for i in countrylist:
    fp="/Users/jiaqitang/Downloads/light/"+i+"22.shp"
    
    c = gpd.read_file(fp)
    c['value'] = c['value'].astype(float)
    c['valuesum']=c['value']
    c['valuemean']=c['value']
    c['valuemedian']=c['value']
    cagg=c.groupby('cst',as_index=False).agg(
    {
         'gdpsum':'sum',    # Sum duration per group
         'gdpmean': "mean",  # get the count of networks
         'gdpmedian': 'median',  # get the first date per group
       'yr':'first',
        'cst_n':'first',
        'ctr_n':'first',
        'ctr':'first',
        
    }
       )
    outfile="/Users/jiaqitang/Downloads/light_gdp_csv/"+i+".csv"
    cagg.to_csv(outfile)
    print(i)
```
